```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
#i. Sum of all elements
#ii. Sum of all elements row-wise
#iii. Sum of all elements column-wise


gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg)

sum1=np.sum(gfg)
print(sum1)

row_sum=np.sum(gfg,axis=1)
print(row_sum)

col_sum=np.sum(gfg,axis=0)
print(col_sum)
     
```

```python
#Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
#i. Sorted array
#ii. Indices of sorted array
#iii. 4 smallest elements
#iv. 5 largest elements

array=np.array([10,52,62,16,16,54,453])
print(array)

sorted_array=np.sort(array)
print(sorted_array)

indices_sort=np.argsort(array)
print(indices_sort)

four_smallest=np.sort(array)[:4]
print(four_smallest)

five_largest=np.sort(array)[-5:]
print(five_largest)
```

```python
#(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
#i. Integer elements only
#ii. Float elements only


array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print(array)

integer_elements = array[array == array.astype(int)]
print(integer_elements)

float_elements=array[array == array.astype(float)]
print(float_elements)
```

```python
#Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting.
#a) Generate your unique sales dataset:

#.Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65+66 = 131 → sales = [131, 181, 231, 281, 331].)
#.Create a NumPy array sales with values [X, X+50, X+100, X+150,X+200].

name = "KA"
X = sum( ord(char) for char in name)
print(X)

sales = np.array([X, X+50, X+100, X+150, X+200])
print(sales)
     
```

```python
#b) Compute your personalized tax rate as ((X % 5) + 5) / 100.
#Use broadcasting to apply this tax rate to each sales value.

tax=((X%5)+5)/100
print(tax)

sales2=sales*(1-tax)
print(sales2)
```

```python
#Adjust sales based on discount:

#If sales < X+100, apply a 5% discount.
#If sales >= X+100, apply a 10% discount.

def adjust_sales(sales, X):
    adjusted_sales = []  # Correct variable name
    
    for sale in sales:
        if sale < X + 100:
            discount = 0.05  # 5% discount
        else:
            discount = 0.10  # 10% discount
        
        adjusted_sales.append(float(sale * (1 - discount)))

    return adjusted_sales


new_sales = adjust_sales(sales, X)
print(f"Adjusted sales: {new_sales}")
```

```python
#d) Expand sales data for multiple weeks:

#Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.
#Increase sales by 2% per week using element-wise broadcasting.

matrix=np.tile(sales,(3,1))
print(matrix,'\n')

weeks=np.array([0,1,2])
weeks=weeks.reshape(-1,1)
increase_sales=matrix*(1+0.02)**weeks
print(increase_sales)
```

```python
#Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy:

#Y = x^2
#Y = sin(x)
#Y= e^x
#Y = log(|x| + 1)
#Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.


x=np.linspace(-10,10,100)
y1=x**2
plt.plot(x,y1)
plt.grid(True)
plt.title("y=x^2")
plt.xlabel("x values")
plt.ylabel("y values")
```

```python
y2=np.sin(x)
plt.plot(x,y2)
plt.grid(True)
plt.title("y=sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
```

```python

y3=np.exp(x)
plt.plot(x,y3)
plt.grid(True)
plt.title("y=e^x")
plt.xlabel("x values")
plt.ylabel("y values")
```

```python

y4=np.log(np.abs(x)+1)
plt.plot(x,y4)
plt.grid(True)
plt.title("y= log(|x| + 1)")
plt.xlabel("x values")
plt.ylabel("y values")
```
