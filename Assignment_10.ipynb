```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))

filtered_tokens=[word for word in word_tokens if word.lower() not in stop_words]
s= ' '.join(filtered_tokens)
print(s)
```

```python
#Q4. Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs Blockchain).

#1.Preprocess and tokenize both texts.
#Calculate:
#2.a. Jaccard Similarity using sets
#b. Cosine Similarity using TfidfVectorizer + cosine_similarity()
#c. Analyze which similarity metric gives better insights in your case.

text1 = """
Internet of Things (IoT) refers to a network of physical devices embedded with sensors and software to collect and exchange data.  
These “smart” devices—from home thermostats to industrial machinery—communicate over the internet, enabling real-time monitoring and control.  
IoT platforms aggregate and analyze this data to optimize performance, predict maintenance needs, and enhance user experiences.  
Security and interoperability are key challenges as device numbers and data volumes continue to grow.
"""

text2 = """
Quantum Computing leverages principles of quantum mechanics—such as superposition and entanglement—to process information in qubits.  
Unlike classical bits, qubits can represent multiple states simultaneously, offering exponential speedups for certain algorithms.  
Potential applications include cryptography, drug discovery, and complex system simulations.  
Hardware stability, error correction, and scalable architectures remain active research areas.
"""
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    return tokens
```

```python
text1_p=preprocess_text(text1)
text2_p=preprocess_text(text2)
```

```python
print(text1_p)
print(text2_p)
```

```python
t1 = set(text1_p)
t2 = set(text2_p)
jaccard = len(t1&t2) / len(t1|t2)
print("Jaccard Similarity:", jaccard)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
tfidf_vec = TfidfVectorizer()
vecs = tfidf_vec.fit_transform([text1,text2])
cos_sim = cosine_similarity(vecs[0:1], vecs[1:2])
print("Cosine Similarity:", cos_sim[0][0])
```

```python
#Q5. Write a short review for a product or service.

#1.Use TextBlob or VADER to find polarity & subjectivity for each review.
#2.Classify reviews as positive, negative, or neutral.
#3.Create a word cloud using the wordcloud library for all positive reviews.

from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

reviews = [
   "An incredibly intuitive gadget that streamlines my entire workflow—absolutely reliable and efficient!",  
   "It handles core tasks smoothly, though I’d love a few more customization options.",  
   "Decently functional; meets basic needs but lacks standout features or flair.",  
   "Disappointingly unstable—frequent crashes and slow load times make it unusable.",  
   "A true standout! Sleek design, rock-solid performance, and exceptional customer support.",  
   "Passable at best—somewhat clunky interface and inconsistent results, but not a total loss.",  
   "Utterly atrocious—never worked from day one, constant errors and zero support. A complete waste!"  
]
```

```python
def classify(compound_score):
    if compound_score>=0.05:
        print("Classification: Positive")
    elif compound_score<=-0.05:
        print("Classification: Negative")
    else:
        print("Classification: Neutral")
```

```python
for review in reviews:
    blob = TextBlob(review)
    print("Polarity:", blob.sentiment.polarity)
    print("Subjectivity:", blob.sentiment.subjectivity)
    sentiment = analyzer.polarity_scores(review)
    print("Sentiment Scores:", sentiment)
    compound_score = sentiment['compound']
    classify(compound_score)
    print()
```

```python
positive=""

for review in reviews:
    sentiment=analyzer.polarity_scores(review)
    if(sentiment['compound']>=0.5):
        positive+=" "+review
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Words of Positive Reviews")
plt.show()
```

```python
#Q6. Choose your own paragraph (~100 words) as training data.

#1.Tokenize text using Tokenizer() from keras.preprocessing.text
#2.Create input sequences and build a simple LSTM or Dense model
#3.Train the model and generate 2–3 new lines of text starting from any seed word you provide.

data = """
In the heart of the bustling city, hidden gardens bloom quietly between towering skyscrapers. 
Colorful flowers sway gently in the breeze, and tiny fountains whisper soothing melodies. 
Visitors, escaping the city's rush, find peace among winding stone paths and shaded benches. 
The scent of jasmine and fresh grass fills the air, mingling with the distant hum of traffic. 
Children chase butterflies, artists sketch, and old friends share stories over cups of steaming tea. 
Each hidden garden is a little sanctuary, a reminder that even amid the chaos, serenity can thrive.
"""
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
total_words = len(tokenizer.word_index) + 1
```

```python
token_list = tokenizer.texts_to_sequences([data])[0]
input_sequences = []

for i in range(1, len(token_list)):
    n_gram_sequence = token_list[:i+1]
    input_sequences.append(n_gram_sequence)
```

```python
padded = pad_sequences(input_sequences, padding='pre')
X = padded[:, :-1]
y = padded[:, -1]
```

```python
model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=10))
model.add(LSTM(50))
model.add(Dense(50, activation='relu'))
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.summary()
```

```python
import numpy as np

def generate_text(seed_text, next_words=15):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=X.shape[1], padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_index = np.argmax(predicted)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text

print("\nGenerated Text:\n")
print(generate_text("breeze"))
```
